```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df.RM.median()
```

```python
df.RM = df.RM.fillna(df.RM.median())
df
```

```python
x = df.drop('MEDV', axis='columns')
x
```

```python
y = df.MEDV
y
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(x, y)
```
