```python
import requests
from bs4 import BeautifulSoup

# 1. نرسل طلب إلى الموقع
url = "http://www.scrapethissite.com/pages/simple/"
response = requests.get(url)

# 2. التحقق من نجاح الطلب
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # 3. البحث عن كل عنصر يمثل دولة
    countries = soup.find_all('div', class_='col-md-4 country')

    print("قائمة الدول وعدد سكانها:\n")

    for country in countries:
        # اسم الدولة موجود في وسم h3 بفئة country-name
        country_name = country.find('h3', class_='country-name').text.strip()

        # عدد السكان موجود في وسم span بفئة country-population
        population = country.find('span', class_='country-population').text.strip()

        print(f"- الدولة: {country_name} | عدد السكان: {population}")

else:
    print("فشل تحميل الصفحة. رمز الاستجابة:", response.status_code)
```
